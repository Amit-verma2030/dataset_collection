```python
x = df[['CustomerID','Age',	'Annual Income (k$)',	'Spending Score (1-100)']]
```

```python

from sklearn.cluster import KMeans
kmeanmodal = KMeans(n_clusters = 2,n_init = 10,random_state = 0)
```

```python
kmeanmodal.fit(x)
```

```python
kmeanmodal.predict(x)
```
